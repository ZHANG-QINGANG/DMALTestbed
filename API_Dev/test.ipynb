```python
import math

#  (Teten)
def saturation_vapor_pressure(T):
    return 0.6108 * math.exp((17.27 * T) / (T + 237.3))  # kPa

# 
def air_enthalpy(T_db, RH, P_atm=101.325):
    #  (kPa)
    P_sat = saturation_vapor_pressure(T_db)
    
    #  (kPa)
    P_v = RH * P_sat
    
    #  (kg/kg)
    W = 0.622 * P_v / (P_atm - P_v)
    
    #  (kJ/kg)
    h = 1.006 * T_db + W * (2501 + 1.805 * T_db)
    
    return h

#  25C  50%
T_db = 25  # C
RH = 0.50  # 

h_r = air_enthalpy(22.7, 0.51)
h_s = air_enthalpy(20.8, 0.58)
h_air_dif = (h_r - h_s) * 1.37 * 1.29
energy_dif = 1.37*1.29*1005*(22.7-20.8)
print(f"Cooling load: {(1770.00 + 1282)/1000} kJ")
print(f": {h_air_dif:.2f} kJ")
print(f": {energy_dif/1000:.2f} kJ")
```

```python
def water_enthalpy_liquid(T):
    """
     (kJ/kg)
    :
    T:  ()
    
    :
    h_liquid:  (kJ/kg)
    """
    c_p_liquid = 4.18  #  (kJ/kgC)
    h_liquid = c_p_liquid * T  
    return h_liquid

def water_enthalpy_steam(T):
    """
     (kJ/kg)
    :
    T:  ()
    
    :
    h_steam:  (kJ/kg)
    """
    h_fg = 2257  #  (kJ/kg)
    c_p_steam = 1.996  #  (kJ/kgC)
    h_steam = h_fg + c_p_steam * T
    return h_steam

# :  100C 
T_water = 100  # 

# 
h_liquid = water_enthalpy_liquid(T_water)
print(f" {T_water}C : {h_liquid:.2f} kJ/kg")
```

```python
# 
import os
import json

#  - 
folder_path = 'coil_data_collection_137'

#  ACU_Return_Temp  ACU_Supply_Temp 
acu_return_temp_list = []
acu_supply_temp_list = []
ITE_load_list = []

data = [0.27, 0.31, 0.74, 1.24, 2.2, 3.02, 3.21, 3.63, 3.71]

# 
for flow in data:
    for filename in os.listdir(folder_path):
        if filename.endswith('.json') and 'FCUAirMassFlowRate_1.37' and str(flow) in filename:
            try:
                file_path = os.path.join(folder_path, filename)
                with open(file_path, 'r') as f:
                    data = json.load(f)
                    acu_return_temp = data.get("ACU_Return_Temp")
                    acu_supply_temp = data.get("ACU_Supply_Temp")
                    ITE_load = data.get("ITE_load")
                    acu_return_temp_list.append(acu_return_temp)
                    acu_supply_temp_list.append(acu_supply_temp)
                    ITE_load_list.append(ITE_load)
            except (IndexError, ValueError, FileNotFoundError, json.JSONDecodeError) as e:
                print(f"Error processing file {filename}: {e}")

cooling_load = (np.array(acu_return_temp_list) - np.array(acu_supply_temp_list)) * 1.37 * 1.29 * 1005
print(cooling_load)
```

```python
from CoolProp.CoolProp import PropsSI
import psychrolib
from loguru import logger
psychrolib.SetUnitSystem(psychrolib.SI)


def get_fluid_property(fluid_name, temperature, property_type):
    try:
        # Convert temperature to Kelvin
        temperature_K = temperature + 273.15  # Assuming input temperature is in Celsius

        # Define property mapping
        property_map = {
            'density': 'D',
            'specific_heat': 'C'
        }

        # Check if the property type is valid
        if property_type not in property_map:
            raise ValueError(f"Invalid property type: {property_type}")

        # Get the property
        prop = PropsSI(property_map[property_type], 'T', temperature_K, 'P', 101325, fluid_name)

        return prop
    except ValueError as e:
        logger.error(f"Error: {e:.2f}")
        return None



# 
T_initial = 19.6  # 
RH_initial = 0.55  # 
T_final = 17.6    # 
RH_final = 0.65    # 

#  psychrolib 
h_initial = psychrolib.GetMoistAirEnthalpy(T_initial, psychrolib.GetHumRatioFromRelHum(T_initial, RH_initial, 101325))
h_final = psychrolib.GetMoistAirEnthalpy(T_final, psychrolib.GetHumRatioFromRelHum(T_final, RH_final, 101325))

rho = get_fluid_property(fluid_name="air",
                         temperature=T_initial,
                         property_type="density")
mass = 1.37 * rho

Cp_in = get_fluid_property(fluid_name="air",
                         temperature=T_initial,
                         property_type="specific_heat")
Cp_out = get_fluid_property(fluid_name="air",
                            temperature=T_final,
                            property_type="specific_heat")


#  1 kg
h_dif = 1 * (h_initial - h_final) * mass #  J/kg
Q_dif = (T_initial * Cp_in - T_final * Cp_out) * mass


print(f": {h_dif:.2f} J/kg."
      f": {Q_dif:.2f} J/kg.")
```
